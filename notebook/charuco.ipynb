```python
import numpy as np
import cv2
import calib
import pickle
import cv2.aruco as aruco
import pathlib
import matplotlib.pyplot as plt
```

```python
square_length = 25/2
marker_length = 18.75/2

aruco_dict = aruco.Dictionary_get(aruco.DICT_6X6_1000)
board = aruco.CharucoBoard_create(8*2, 8*2, square_length, marker_length, aruco_dict)
```

```python
calib.save_board("board.pkl", board, aruco.DICT_6X6_1000)

board == calib.load_board("board.pkl")
```

```python
size = (756,756)
img = board.draw(size)
# img2 = calib.load_board("board.pkl")[0].draw(size)
```

```python
plt.imshow(img)
plt.figure()
# plt.imshow(img2)
cv2.imwrite("out_board.png", img)
```
